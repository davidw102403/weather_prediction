Since most days had no precip, it is reasonable to replace null values with zero. Likewise for snow and snow_depth

```python
core_weather["precip"] = core_weather["precip"].fillna(0)
core_weather["snow"] = core_weather["snow"].fillna(0)
core_weather["snow_depth"] = core_weather["snow_depth"].fillna(0)
core_weather
```

Use a forward fill to replace null temp values

```python
# cut off the NAN daterange for temp
core_weather = core_weather.dropna(subset=["temp_min", "temp_max"])
core_weather
```

```python
# forward fill remaining nan values
core_weather["temp_max"] = core_weather["temp_max"].fillna(method="ffill")
core_weather["temp_min"] = core_weather["temp_min"].fillna(method="ffill")
```

## Machine Learning 

The goal of the model will be to predict tomorrow's max temp using historical data

```python
# create new column target from copy of temp_max, shifted one row up
core_weather["target"] = core_weather.shift(-1)["temp_max"].copy()
```

```python
core_weather
```

```python
# delete the last row with NaN
core_weather = core_weather.iloc[:-1,:].copy()
core_weather
```

```python
from sklearn.linear_model import Ridge

# initialize model
reg = Ridge(alpha=.1)
```

```python
predictors = ["precip", "snow", "snow_depth", "temp_max", "temp_min"]
```

```python
# split the data for training and testing
# it is important that the training data comes before the testing data
train = core_weather.loc[:"2020-12-31"]
test = core_weather.loc["2021-01-01":]
test[predictors]
```

```python
# train the model
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
# evaluate the model
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["target"], predictions)
```

The model is off an average of 6.46 degrees

## Evaluating the model

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
combined
```

```python
combined.plot()
```

predictions seem to be a little more conservative than the actual

```python
reg.coef_
```

## Creating a fuction to make predictions

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)

    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]

    return error, combined
    
```

## Improving the model

add more predictors

```python
# create rolling average for max temp for each month
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
core_weather
```

```python
# ratio between monthly max temp average and daily max temp
core_weather["month_day_max"] = core_weather["month_max"] / core_weather["temp_max"]
```

```python
# ratio between max temp and min temp
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
```

```python
predictors = ["precip", "snow", "snow_depth", "temp_max", "temp_min", "month_day_max", "max_min"]
```

```python
# remove nan rows
core_weather = core_weather.iloc[30:,:].copy()
core_weather
```

```python
# remove extreme values
import numpy as np
core_weather = core_weather[~core_weather.isin([np.inf, -np.inf]).any(axis=1)].copy()
core_weather
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
error
```

```python
combined.plot()
```

adding more predictors

```python
# monthly average using only previous data 
core_weather["monthly_avg"] = core_weather.groupby(core_weather.index.month)["temp_max"].expanding().mean().reset_index(0, drop=True)
core_weather
```

```python
# yearly average using only previous data
core_weather["day_of_year_avg"] = core_weather["temp_max"].groupby(core_weather.index.day_of_year).apply(lambda x: x.expanding().mean()).reset_index(0, drop=True)
```

```python
predictors = ["precip", "snow", "snow_depth", "temp_max", "temp_min", "month_day_max", "max_min", "monthly_avg", "day_of_year_avg"]
```

```python
error, combined = create_predictions(predictors, core_weather, reg)
error
```

improved from 6.41 to 6.08
